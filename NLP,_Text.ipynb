```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag

# Download required NLTK data
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')

def pos_tag_sentence(text):
    """
    Function to tokenize and POS tag a given sentence.
    Returns a list of (word, tag) tuples.
    """
    # Step 1: Tokenize the text into words
    tokens = word_tokenize(text)

    # Step 2: Apply POS tagging
    tagged = pos_tag(tokens)

    return tagged

# Example usage
input_text = "Text filtering systems use NLP to classify documents."
result = pos_tag_sentence(input_text)

print("=" * 50)
print("FIG 3: PYTHON CODE FOR POS TAGGING")
print("=" * 50)
print("\n--- Function Code ---")
print("def pos_tag_sentence(text):")
print("    tokens = word_tokenize(text)")
print("    tagged = pos_tag(tokens)")
print("    return tagged\n")
print(f"Input:  {input_text}")
print(f"Output: {result}")
```

```python
import nltk
from nltk import ngrams, FreqDist
from nltk.tokenize import word_tokenize
from collections import defaultdict

nltk.download('punkt', quiet=True)

# Training corpus
corpus = """
Natural language processing is fascinating.
Language models predict the next word.
Natural language understanding is important.
Text classification filters spam.
Spam filtering uses NLP techniques.
"""

# Preprocess
tokens = word_tokenize(corpus.lower())
tokens = [t for t in tokens if t.isalpha()]

# Build bigram model
bigrams = list(ngrams(tokens, 2))
freq_dist = FreqDist(bigrams)

# Build prediction dictionary
predictions = defaultdict(list)
for w1, w2 in bigrams:
    predictions[w1].append(w2)

def predict_next_word(word):
    word = word.lower()
    if word in predictions:
        next_words = predictions[word]
        fd = FreqDist(next_words)
        most_common = fd.most_common(3)
        return most_common
    return []

print("=" * 50)
print("FIG 4: LANGUAGE PREDICTION EXAMPLE")
print("=" * 50)
print("Training Corpus: 6 sentences on NLP/Spam Filtering\n")

test_words = ['natural', 'language', 'spam', 'text']
for word in test_words:
    preds = predict_next_word(word)
    print(f"Input: '{word}'")
    if preds:
        for next_w, count in preds:
            print(f"  → Predicted: '{next_w}' (frequency: {count})")
    else:
        print("  → No prediction available")
    print()
```

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.tag import pos_tag
from nltk import FreqDist

nltk.download('punkt', quiet=True)
nltk.download('averaged_perceptron_tagger', quiet=True)
nltk.download('maxent_ne_chunker_tab', quiet=True) # Added to resolve LookupError

text = """Natural Language Processing is a subfield of artificial intelligence.
It enables computers to understand and generate human language.
NLP powers applications like spam filtering and machine translation."""

print("=" * 55)
print("FIG 5: SENTENCE ANALYSIS OUTPUT")
print("=" * 55)

# 1. Sentence Tokenization
sentences = sent_tokenize(text)
print(f"\n1. SENTENCE COUNT: {len(sentences)}")
for i, sent in enumerate(sentences, 1):
    print(f"   [{i}] {sent}")

# 2. Word Tokenization & POS per sentence
print(f"\n2. TOKEN & POS ANALYSIS PER SENTENCE:")
for i, sent in enumerate(sentences, 1):
    tokens = word_tokenize(sent)
    tagged = pos_tag(tokens)
    print(f"\n   Sentence {i}: {sent}")
    print(f"   Word count: {len(tokens)}")
    print(f"   POS Tags: {tagged}")

# 3. Overall statistics
all_tokens = word_tokenize(text)
all_tagged = pos_tag(all_tokens)
pos_counts = FreqDist(tag for word, tag in all_tagged)

print(f"\n3. OVERALL STATISTICS:")
print(f"   Total words: {len(all_tokens)}")
print(f"   Unique words: {len(set(w.lower() for w in all_tokens if w.isalpha()))}")
print(f"\n   POS Frequency Distribution:")
for tag, count in pos_counts.most_common():
    print(f"      {tag}: {count}")
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag

nltk.download('punkt', quiet=True)
nltk.download('averaged_perceptron_tagger', quiet=True)

def sequence_label(sentence):
    """
    Sequence Labeling: Assigns a label (POS tag) to each token
    in the input sequence.
    """
    tokens = word_tokenize(sentence)
    labeled_sequence = pos_tag(tokens)
    return labeled_sequence

# Example usage
sentence = "Natural language processing improves spam filtering."
result = sequence_label(sentence)

print("=" * 55)
print("FIG 1: SEQUENCE LABELING CODE")
print("=" * 55)
print(f"Input Sentence: {sentence}\n")
print("Labeled Sequence (Token, Tag):")
for word, tag in result:
    print(f"  {word:<15} → {tag}")
```

```python
import nltk
from nltk.tag import hmm
from nltk.tokenize import word_tokenize

nltk.download('punkt', quiet=True)

# Training data: list of (token, tag) sequences
train_data = [
    [('the', 'DT'), ('cat', 'NN'), ('sleeps', 'VBZ')],
    [('the', 'DT'), ('dog', 'NN'), ('runs', 'VBZ')],
    [('a', 'DT'), ('fox', 'NN'), ('jumps', 'VBZ')],
    [('the', 'DT'), ('spam', 'NN'), ('filters', 'VBZ')],
    [('natural', 'JJ'), ('language', 'NN'), ('processes', 'VBZ')],
]

# Test sentence
test_sentence = "the spam filters"
test_tokens = word_tokenize(test_sentence)

# Train HMM tagger
trainer = hmm.HiddenMarkovModelTrainer()
hmm_tagger = trainer.train(train_data)

# Predict tags
predicted = hmm_tagger.tag(test_tokens)

print("=" * 55)
print("FIG 2: HIDDEN MARKOV MODEL OUTPUT")
print("=" * 55)
print(f"Training sentences: {len(train_data)}")
print(f"Test sentence: '{test_sentence}'\n")
print("HMM Predicted Sequence:")
for word, tag in predicted:
    print(f"  {word:<10} → {tag}")
```

```python
import nltk
from nltk import ne_chunk, pos_tag, word_tokenize
from nltk.tree import Tree

nltk.download('punkt', quiet=True)
nltk.download('averaged_perceptron_tagger', quiet=True)
nltk.download('maxent_ne_chunker', quiet=True)
nltk.download('words', quiet=True)

sentence = "Apple Inc. is planning to open a new store in London by January 2025."

# Tokenize and POS tag
tokens = word_tokenize(sentence)
tagged = pos_tag(tokens)

# Named Entity Recognition
tree = ne_chunk(tagged)

print("=" * 55)
print("FIG 3: NAMED ENTITY RECOGNITION EXAMPLE")
print("=" * 55)
print(f"Sentence: {sentence}\n")
print("Named Entities Found:")

entities = []
for chunk in tree:
    if isinstance(chunk, Tree):
        entity_name = " ".join(token for token, pos in chunk.leaves())
        entity_type = chunk.label()
        entities.append((entity_name, entity_type))
        print(f"  {entity_name:<20} → {entity_type}")

if not entities:
    print("  No named entities found.")
```

```python
import nltk
from nltk import ne_chunk, pos_tag, word_tokenize
from nltk.tree import Tree

nltk.download('punkt', quiet=True)
nltk.download('averaged_perceptron_tagger', quiet=True)
nltk.download('maxent_ne_chunker', quiet=True)
nltk.download('words', quiet=True)

sentences = [
    "Google develops AI tools in California.",
    "Dr. Smith works at the University of Cambridge.",
    "Email spam filtering protects users daily."
]

print("=" * 65)
print("FIG 4: SENTENCE LABELING RESULTS")
print("=" * 65)

for i, sent in enumerate(sentences, 1):
    tokens = word_tokenize(sent)
    tagged = pos_tag(tokens)
    tree = ne_chunk(tagged)

    # Extract entities
    entities = []
    for chunk in tree:
        if isinstance(chunk, Tree):
            name = " ".join(t for t, p in chunk.leaves())
            entities.append(f"{name}({chunk.label()})")

    print(f"\nSentence {i}: {sent}")
    print(f"  POS Tags:     {tagged}")
    print(f"  Entities:     {', '.join(entities) if entities else 'None'}")
    print(f"  Token Count:  {len(tokens)}")
```

```python
import spacy

# Load the small English model (includes dependency parser)
nlp = spacy.load("en_core_web_sm")

sentences = [
    "The administrator updated student records",
    "Machine learning improves language processing"
]

print("=" * 70)
print("ASSIGNMENT 1: DEPENDENCY STRUCTURE ANALYSIS")
print("=" * 70)

for sent in sentences:
    doc = nlp(sent)
    print(f"\nSentence: '{sent}'")
    print(f"{'Token':<18} {'Dependency':<14} {'Head':<14} {'Explanation'}")
    print("-" * 70)
    for token in doc:
        print(f"{token.text:<18} {token.dep_:<14} {token.head.text:<14} {spacy.explain(token.dep_)}")
```

```python
import spacy

# Load the medium English model (includes word vectors for similarity)
nlp = spacy.load("en_core_web_md")

# Pair 1: Semantically similar sentences
similar_pair = (
    "Natural language processing helps computers understand text",
    "NLP enables machines to comprehend human language"
)

# Pair 2: Unrelated sentences
unrelated_pair = (
    "The weather is sunny today",
    "Machine learning improves spam filtering"
)

print("=" * 70)
print("ASSIGNMENT 2: SEMANTIC SIMILARITY COMPARISON")
print("=" * 70)

for label, pair in [("SIMILAR", similar_pair), ("UNRELATED", unrelated_pair)]:
    doc1 = nlp(pair[0])
    doc2 = nlp(pair[1])
    score = doc1.similarity(doc2)

    print(f"\n{'='*50}")
    print(f"Category: {label}")
    print(f"Similarity Score: {score:.4f}")
    print(f"  Sentence A: {pair[0]}")
    print(f"  Sentence B: {pair[1]}")
```

```python
import re
import random

class AcademicChatbot:
    def __init__(self):
        self.name = "AcademicBot"

        # Greeting responses
        self.greetings = [
            "Hello! I'm your Academic Assistant. How can I help you today?",
            "Hi there! Ready to assist with your academic queries.",
            "Hey! Ask me about CAT schedules or unit registration."
        ]

        # CAT schedule knowledge base
        self.cat_info = {
            "when": "CATs are scheduled from Week 6 to Week 8. Check the student portal for your exact timetable.",
            "where": "CATs take place in your assigned lecture halls. Room allocations are posted on the notice board and portal.",
            "format": "CATs include multiple-choice and short-answer questions. Duration is usually 1 to 2 hours.",
            "general": "I can tell you about CAT dates, venues, and format. What would you like to know?"
        }

        # Unit registration knowledge base
        self.reg_info = {
            "deadline": "Unit registration closes on Friday of Week 2. Late registration attracts a penalty fee.",
            "how": "Log into the student portal → Academics → Unit Registration → Select your units → Click Submit.",
            "add": "You can add units during the add/drop period (Week 1 to Week 3) via the student portal.",
            "drop": "Units can be dropped before the end of Week 3. After that, you must complete the unit.",
            "general": "I can help with registration deadlines, steps, adding or dropping units. What do you need?"
        }

        self.farewells = [
            "Goodbye! Have a productive semester!",
            "See you later! Feel free to return anytime.",
            "Bye! Best of luck with your studies."
        ]

        self.unknown = [
            "I'm not sure I understand. Try asking about CAT schedules or unit registration.",
            "Could you rephrase? I can help with greetings, CAT info, registration, or exit commands.",
            "I only handle academic queries. Type 'help' to see what I can do."
        ]

    def get_intent(self, user_input):
        """Classify user input into an intent using keyword matching."""
        text = user_input.lower().strip()

        # Exit patterns
        if re.search(r'\b(bye|exit|quit|goodbye|see you)\b', text):
            return "exit"

        # Greeting patterns
        if re.search(r'\b(hello|hi|hey|greetings|morning|afternoon)\b', text):
            return "greeting"

        # Help
        if re.search(r'\b(help|what can you do|options)\b', text):
            return "help"

        # CAT schedule patterns
        if re.search(r'\b(cat|cats|continuous assessment|test)\b', text):
            if re.search(r'\b(when|date|time|schedule|week)\b', text):
                return "cat_when"
            elif re.search(r'\b(where|venue|room|hall|place)\b', text):
                return "cat_where"
            elif re.search(r'\b(format|type|structure|duration|long)\b', text):
                return "cat_format"
            else:
                return "cat_general"

        # Registration patterns
        if re.search(r'\b(register|registration|unit|course|enroll|enrollment)\b', text):
            if re.search(r'\b(deadline|when|close|last day|due)\b', text):
                return "reg_deadline"
            elif re.search(r'\b(how|steps|process|way|procedure)\b', text):
                return "reg_how"
            elif re.search(r'\b(add|extra|more|additional)\b', text):
                return "reg_add"
            elif re.search(r'\b(drop|remove|withdraw|delete)\b', text):
                return "reg_drop"
            else:
                return "reg_general"

        return "unknown"

    def respond(self, user_input):
        """Generate a response based on detected intent."""
        intent = self.get_intent(user_input)

        if intent == "exit":
            return random.choice(self.farewells), False

        elif intent == "greeting":
            return random.choice(self.greetings), True

        elif intent == "help":
            help_text = (
                "Here is what I can do:\n"
                "  1. Greetings — Just say hello!\n"
                "  2. CAT schedules — Ask 'When are CATs?' or 'Where is my CAT?'\n"
                "  3. Unit registration — Ask 'How do I register?' or 'Registration deadline?'\n"
                "  4. Exit — Say 'bye' or 'exit' to end our chat."
            )
            return help_text, True

        elif intent == "cat_when":
            return self.cat_info["when"], True
        elif intent == "cat_where":
            return self.cat_info["where"], True
        elif intent == "cat_format":
            return self.cat_info["format"], True
        elif intent == "cat_general":
            return self.cat_info["general"], True

        elif intent == "reg_deadline":
            return self.reg_info["deadline"], True
        elif intent == "reg_how":
            return self.reg_info["how"], True
        elif intent == "reg_add":
            return self.reg_info["add"], True
        elif intent == "reg_drop":
            return self.reg_info["drop"], True
        elif intent == "reg_general":
            return self.reg_info["general"], True

        else:
            return random.choice(self.unknown), True

    def run(self):
        """Start the chatbot conversation loop."""
        print("=" * 60)
        print("  STUDENT ACADEMIC ASSISTANT CHATBOT")
        print("=" * 60)
        print("Type 'help' for options or 'exit' to quit.\n")

        active = True
        while active:
            try:
                user_input = input("You: ").strip()
                if not user_input:
                    continue

                response, active = self.respond(user_input)
                print(f"{self.name}: {response}")

                if not active:
                    print("\n[Chat session ended]")

            except (KeyboardInterrupt, EOFError):
                print(f"\n{self.name}: {random.choice(self.farewells)}")
                break


# Run the chatbot
if __name__ == "__main__":
    bot = AcademicChatbot()
    bot.run()
```
