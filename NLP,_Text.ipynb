```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag

# Download required NLTK data
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')

def pos_tag_sentence(text):
    """
    Function to tokenize and POS tag a given sentence.
    Returns a list of (word, tag) tuples.
    """
    # Step 1: Tokenize the text into words
    tokens = word_tokenize(text)

    # Step 2: Apply POS tagging
    tagged = pos_tag(tokens)

    return tagged

# Example usage
input_text = "Text filtering systems use NLP to classify documents."
result = pos_tag_sentence(input_text)

print("=" * 50)
print("FIG 3: PYTHON CODE FOR POS TAGGING")
print("=" * 50)
print("\n--- Function Code ---")
print("def pos_tag_sentence(text):")
print("    tokens = word_tokenize(text)")
print("    tagged = pos_tag(tokens)")
print("    return tagged\n")
print(f"Input:  {input_text}")
print(f"Output: {result}")
```

```python
import nltk
from nltk import ngrams, FreqDist
from nltk.tokenize import word_tokenize
from collections import defaultdict

nltk.download('punkt', quiet=True)

# Training corpus
corpus = """
Natural language processing is fascinating.
Language models predict the next word.
Natural language understanding is important.
Text classification filters spam.
Spam filtering uses NLP techniques.
"""

# Preprocess
tokens = word_tokenize(corpus.lower())
tokens = [t for t in tokens if t.isalpha()]

# Build bigram model
bigrams = list(ngrams(tokens, 2))
freq_dist = FreqDist(bigrams)

# Build prediction dictionary
predictions = defaultdict(list)
for w1, w2 in bigrams:
    predictions[w1].append(w2)

def predict_next_word(word):
    word = word.lower()
    if word in predictions:
        next_words = predictions[word]
        fd = FreqDist(next_words)
        most_common = fd.most_common(3)
        return most_common
    return []

print("=" * 50)
print("FIG 4: LANGUAGE PREDICTION EXAMPLE")
print("=" * 50)
print("Training Corpus: 6 sentences on NLP/Spam Filtering\n")

test_words = ['natural', 'language', 'spam', 'text']
for word in test_words:
    preds = predict_next_word(word)
    print(f"Input: '{word}'")
    if preds:
        for next_w, count in preds:
            print(f"  → Predicted: '{next_w}' (frequency: {count})")
    else:
        print("  → No prediction available")
    print()
```

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.tag import pos_tag
from nltk import FreqDist

nltk.download('punkt', quiet=True)
nltk.download('averaged_perceptron_tagger', quiet=True)
nltk.download('maxent_ne_chunker_tab', quiet=True) # Added to resolve LookupError

text = """Natural Language Processing is a subfield of artificial intelligence.
It enables computers to understand and generate human language.
NLP powers applications like spam filtering and machine translation."""

print("=" * 55)
print("FIG 5: SENTENCE ANALYSIS OUTPUT")
print("=" * 55)

# 1. Sentence Tokenization
sentences = sent_tokenize(text)
print(f"\n1. SENTENCE COUNT: {len(sentences)}")
for i, sent in enumerate(sentences, 1):
    print(f"   [{i}] {sent}")

# 2. Word Tokenization & POS per sentence
print(f"\n2. TOKEN & POS ANALYSIS PER SENTENCE:")
for i, sent in enumerate(sentences, 1):
    tokens = word_tokenize(sent)
    tagged = pos_tag(tokens)
    print(f"\n   Sentence {i}: {sent}")
    print(f"   Word count: {len(tokens)}")
    print(f"   POS Tags: {tagged}")

# 3. Overall statistics
all_tokens = word_tokenize(text)
all_tagged = pos_tag(all_tokens)
pos_counts = FreqDist(tag for word, tag in all_tagged)

print(f"\n3. OVERALL STATISTICS:")
print(f"   Total words: {len(all_tokens)}")
print(f"   Unique words: {len(set(w.lower() for w in all_tokens if w.isalpha()))}")
print(f"\n   POS Frequency Distribution:")
for tag, count in pos_counts.most_common():
    print(f"      {tag}: {count}")
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag

nltk.download('punkt', quiet=True)
nltk.download('averaged_perceptron_tagger', quiet=True)

def sequence_label(sentence):
    """
    Sequence Labeling: Assigns a label (POS tag) to each token
    in the input sequence.
    """
    tokens = word_tokenize(sentence)
    labeled_sequence = pos_tag(tokens)
    return labeled_sequence

# Example usage
sentence = "Natural language processing improves spam filtering."
result = sequence_label(sentence)

print("=" * 55)
print("FIG 1: SEQUENCE LABELING CODE")
print("=" * 55)
print(f"Input Sentence: {sentence}\n")
print("Labeled Sequence (Token, Tag):")
for word, tag in result:
    print(f"  {word:<15} → {tag}")
```

```python
import nltk
from nltk.tag import hmm
from nltk.tokenize import word_tokenize

nltk.download('punkt', quiet=True)

# Training data: list of (token, tag) sequences
train_data = [
    [('the', 'DT'), ('cat', 'NN'), ('sleeps', 'VBZ')],
    [('the', 'DT'), ('dog', 'NN'), ('runs', 'VBZ')],
    [('a', 'DT'), ('fox', 'NN'), ('jumps', 'VBZ')],
    [('the', 'DT'), ('spam', 'NN'), ('filters', 'VBZ')],
    [('natural', 'JJ'), ('language', 'NN'), ('processes', 'VBZ')],
]

# Test sentence
test_sentence = "the spam filters"
test_tokens = word_tokenize(test_sentence)

# Train HMM tagger
trainer = hmm.HiddenMarkovModelTrainer()
hmm_tagger = trainer.train(train_data)

# Predict tags
predicted = hmm_tagger.tag(test_tokens)

print("=" * 55)
print("FIG 2: HIDDEN MARKOV MODEL OUTPUT")
print("=" * 55)
print(f"Training sentences: {len(train_data)}")
print(f"Test sentence: '{test_sentence}'\n")
print("HMM Predicted Sequence:")
for word, tag in predicted:
    print(f"  {word:<10} → {tag}")
```

```python
import nltk
from nltk import ne_chunk, pos_tag, word_tokenize
from nltk.tree import Tree

nltk.download('punkt', quiet=True)
nltk.download('averaged_perceptron_tagger', quiet=True)
nltk.download('maxent_ne_chunker', quiet=True)
nltk.download('words', quiet=True)

sentence = "Apple Inc. is planning to open a new store in London by January 2025."

# Tokenize and POS tag
tokens = word_tokenize(sentence)
tagged = pos_tag(tokens)

# Named Entity Recognition
tree = ne_chunk(tagged)

print("=" * 55)
print("FIG 3: NAMED ENTITY RECOGNITION EXAMPLE")
print("=" * 55)
print(f"Sentence: {sentence}\n")
print("Named Entities Found:")

entities = []
for chunk in tree:
    if isinstance(chunk, Tree):
        entity_name = " ".join(token for token, pos in chunk.leaves())
        entity_type = chunk.label()
        entities.append((entity_name, entity_type))
        print(f"  {entity_name:<20} → {entity_type}")

if not entities:
    print("  No named entities found.")
```

```python
import nltk
from nltk import ne_chunk, pos_tag, word_tokenize
from nltk.tree import Tree

nltk.download('punkt', quiet=True)
nltk.download('averaged_perceptron_tagger', quiet=True)
nltk.download('maxent_ne_chunker', quiet=True)
nltk.download('words', quiet=True)

sentences = [
    "Google develops AI tools in California.",
    "Dr. Smith works at the University of Cambridge.",
    "Email spam filtering protects users daily."
]

print("=" * 65)
print("FIG 4: SENTENCE LABELING RESULTS")
print("=" * 65)

for i, sent in enumerate(sentences, 1):
    tokens = word_tokenize(sent)
    tagged = pos_tag(tokens)
    tree = ne_chunk(tagged)

    # Extract entities
    entities = []
    for chunk in tree:
        if isinstance(chunk, Tree):
            name = " ".join(t for t, p in chunk.leaves())
            entities.append(f"{name}({chunk.label()})")

    print(f"\nSentence {i}: {sent}")
    print(f"  POS Tags:     {tagged}")
    print(f"  Entities:     {', '.join(entities) if entities else 'None'}")
    print(f"  Token Count:  {len(tokens)}")
```

```python
import re
import random

class AcademicChatbot:
    def __init__(self):
        self.name = "AcademicBot"

        # Greeting responses
        self.greetings = [
            "Hello! I'm your Academic Assistant. How can I help you today?",
            "Hi there! Ready to assist with your academic queries.",
            "Hey! Ask me about CAT schedules or unit registration."
        ]

        # CAT schedule knowledge base
        self.cat_info = {
            "when": "CATs are scheduled from Week 6 to Week 8. Check the student portal for your exact timetable.",
            "where": "CATs take place in your assigned lecture halls. Room allocations are posted on the notice board and portal.",
            "format": "CATs include multiple-choice and short-answer questions. Duration is usually 1 to 2 hours.",
            "general": "I can tell you about CAT dates, venues, and format. What would you like to know?"
        }

        # Unit registration knowledge base
        self.reg_info = {
            "deadline": "Unit registration closes on Friday of Week 2. Late registration attracts a penalty fee.",
            "how": "Log into the student portal → Academics → Unit Registration → Select your units → Click Submit.",
            "add": "You can add units during the add/drop period (Week 1 to Week 3) via the student portal.",
            "drop": "Units can be dropped before the end of Week 3. After that, you must complete the unit.",
            "general": "I can help with registration deadlines, steps, adding or dropping units. What do you need?"
        }

        self.farewells = [
            "Goodbye! Have a productive semester!",
            "See you later! Feel free to return anytime.",
            "Bye! Best of luck with your studies."
        ]

        self.unknown = [
            "I'm not sure I understand. Try asking about CAT schedules or unit registration.",
            "Could you rephrase? I can help with greetings, CAT info, registration, or exit commands.",
            "I only handle academic queries. Type 'help' to see what I can do."
        ]

    def get_intent(self, user_input):
        """Classify user input into an intent using keyword matching."""
        text = user_input.lower().strip()

        # Exit patterns
        if re.search(r'\b(bye|exit|quit|goodbye|see you)\b', text):
            return "exit"

        # Greeting patterns
        if re.search(r'\b(hello|hi|hey|greetings|morning|afternoon)\b', text):
            return "greeting"

        # Help
        if re.search(r'\b(help|what can you do|options)\b', text):
            return "help"

        # CAT schedule patterns
        if re.search(r'\b(cat|cats|continuous assessment|test)\b', text):
            if re.search(r'\b(when|date|time|schedule|week)\b', text):
                return "cat_when"
            elif re.search(r'\b(where|venue|room|hall|place)\b', text):
                return "cat_where"
            elif re.search(r'\b(format|type|structure|duration|long)\b', text):
                return "cat_format"
            else:
                return "cat_general"

        # Registration patterns
        if re.search(r'\b(register|registration|unit|course|enroll|enrollment)\b', text):
            if re.search(r'\b(deadline|when|close|last day|due)\b', text):
                return "reg_deadline"
            elif re.search(r'\b(how|steps|process|way|procedure)\b', text):
                return "reg_how"
            elif re.search(r'\b(add|extra|more|additional)\b', text):
                return "reg_add"
            elif re.search(r'\b(drop|remove|withdraw|delete)\b', text):
                return "reg_drop"
            else:
                return "reg_general"

        return "unknown"

    def respond(self, user_input):
        """Generate a response based on detected intent."""
        intent = self.get_intent(user_input)

        if intent == "exit":
            return random.choice(self.farewells), False

        elif intent == "greeting":
            return random.choice(self.greetings), True

        elif intent == "help":
            help_text = (
                "Here is what I can do:\n"
                "  1. Greetings — Just say hello!\n"
                "  2. CAT schedules — Ask 'When are CATs?' or 'Where is my CAT?'\n"
                "  3. Unit registration — Ask 'How do I register?' or 'Registration deadline?'\n"
                "  4. Exit — Say 'bye' or 'exit' to end our chat."
            )
            return help_text, True

        elif intent == "cat_when":
            return self.cat_info["when"], True
        elif intent == "cat_where":
            return self.cat_info["where"], True
        elif intent == "cat_format":
            return self.cat_info["format"], True
        elif intent == "cat_general":
            return self.cat_info["general"], True

        elif intent == "reg_deadline":
            return self.reg_info["deadline"], True
        elif intent == "reg_how":
            return self.reg_info["how"], True
        elif intent == "reg_add":
            return self.reg_info["add"], True
        elif intent == "reg_drop":
            return self.reg_info["drop"], True
        elif intent == "reg_general":
            return self.reg_info["general"], True

        else:
            return random.choice(self.unknown), True

    def run(self):
        """Start the chatbot conversation loop."""
        print("=" * 60)
        print("  STUDENT ACADEMIC ASSISTANT CHATBOT")
        print("=" * 60)
        print("Type 'help' for options or 'exit' to quit.\n")

        active = True
        while active:
            try:
                user_input = input("You: ").strip()
                if not user_input:
                    continue

                response, active = self.respond(user_input)
                print(f"{self.name}: {response}")

                if not active:
                    print("\n[Chat session ended]")

            except (KeyboardInterrupt, EOFError):
                print(f"\n{self.name}: {random.choice(self.farewells)}")
                break


# Run the chatbot
if __name__ == "__main__":
    bot = AcademicChatbot()
    bot.run()
```

### Training a Word2Vec Model

To effectively demonstrate the capabilities of Word2Vec, we'll use a slightly larger and more diverse set of sentences. This will allow the model to learn more robust word embeddings.

```python
from gensim.models import Word2Vec

# Expanded dataset for better word embeddings
expanded_sentences = [
    ["natural", "language", "processing", "is", "a", "field", "of", "artificial", "intelligence"],
    ["nlp", "is", "used", "for", "text", "analysis", "and", "understanding"],
    ["machine", "learning", "algorithms", "are", "fundamental", "to", "modern", "nlp"],
    ["word2vec", "is", "a", "popular", "technique", "for", "word", "embeddings"],
    ["deep", "learning", "has", "revolutionized", "many", "nlp", "tasks"],
    ["the", "cat", "sat", "on", "the", "mat"],
    ["the", "dog", "chased", "the", "cat"],
    ["i", "love", "reading", "books", "about", "science"],
    ["data", "science", "is", "an", "interdisciplinary", "field"],
    ["artificial", "intelligence", "is", "transforming", "various", "industries"]
]

# Train the Word2Vec model with the expanded dataset
# vector_size: Dimensionality of the word vectors
# window: Maximum distance between the current and predicted word within a sentence
# min_count: Ignores all words with total frequency lower than this
model_expanded = Word2Vec(expanded_sentences,
                          vector_size=100,
                          window=5,
                          min_count=1,
                          sg=0) # sg=0 for CBOW, sg=1 for Skip-gram

print("Word2Vec model trained successfully!")
```

### Finding Similar Words

One of the key applications of Word2Vec is to find words that are semantically similar. This is done by calculating the cosine similarity between word vectors.

```python
print("=" * 50)
print("SIMILAR WORDS TO 'nlp'")
print("=" * 50)
# Find words most similar to 'nlp'
if "nlp" in model_expanded.wv:
    similar_words_nlp = model_expanded.wv.most_similar("nlp", topn=5)
    for word, score in similar_words_nlp:
        print(f"{word}: {score:.4f}")
else:
    print("'nlp' not in vocabulary")

print("\n" + "=" * 50)
print("SIMILAR WORDS TO 'cat'")
print("=" * 50)
# Find words most similar to 'cat'
if "cat" in model_expanded.wv:
    similar_words_cat = model_expanded.wv.most_similar("cat", topn=3)
    for word, score in similar_words_cat:
        print(f"{word}: {score:.4f}")
else:
    print("'cat' not in vocabulary")
```

## Practical Task 2: Similarity Analysis

In this task, we will explore the semantic relationships learned by our Word2Vec model by:
1.  Selecting 10 words from our corpus.
2.  Finding words most similar to each of the selected words.
3.  Calculating the similarity score between specific word pairs.

## Practical Task 3: Comparison Report - One-Hot Encoding vs. Word Embeddings

This report compares two fundamental techniques for representing categorical data, specifically words in Natural Language Processing (NLP): **One-Hot Encoding** and **Word Embeddings**.

| Feature               | One-Hot Encoding                                      | Word Embeddings (e.g., Word2Vec, GloVe, FastText)       |
| :-------------------- | :---------------------------------------------------- | :------------------------------------------------------ |
| **Size (Dimensionality)** | - High-dimensional (vocabulary size)<br>- Sparse vectors (mostly zeros) | - Lower-dimensional (e.g., 50-300)<br>- Dense vectors (mostly non-zero values) |
| **Meaning Representation** | - No semantic meaning captured.<br>- Orthogonal vectors (no inherent relationship between words). | - Captures semantic and syntactic relationships.<br>- Words with similar meanings are close in the vector space. |
| **Efficiency**        | - Computationally inefficient for large vocabularies (due to sparsity and high dimensionality).<br>- Requires more memory. | - More computationally efficient for large vocabularies (due to dense, lower-dimensional vectors).<br>- Requires less memory for representation once trained. |
| **Semantic Understanding** | - Treats each word as a unique, independent entity.<br>- Cannot understand synonyms, antonyms, or contextual meaning. | - Understands words based on their context.<br>- Can infer relationships like 'king' - 'man' + 'woman' = 'queen'. |

### Conclusion

**One-Hot Encoding** is simple to implement and understand, making it suitable for very small vocabularies or when dealing with entirely independent categorical features where no semantic relationship is expected. However, its major drawbacks are the high dimensionality, sparsity, and inability to capture any semantic meaning or relationships between words.

**Word Embeddings**, on the other hand, overcome these limitations by representing words as dense, low-dimensional vectors that capture rich semantic and syntactic relationships. This makes them significantly more powerful for tasks requiring an understanding of language nuances, such as sentiment analysis, machine translation, and question answering. While training word embedding models can be computationally intensive, their resulting representations are far more efficient and effective for downstream NLP tasks, leading to better performance and more robust models. For most modern NLP applications, word embeddings are the preferred method of word representation.

## Advanced Practical Task: Word2Vec on a Larger Dataset

For this advanced task, we will train a Word2Vec model on a more substantial dataset, explore its semantic relationships, and visualize word clusters. This will provide a better understanding of how word embeddings capture meaning from context.

### Step 1: Download and Prepare a Larger Corpus (Gutenberg)

We will use the `gutenberg` corpus from NLTK, which contains a collection of classic literature, suitable for training word embedding models.

```python
import nltk
from gensim.models import Word2Vec

# Download the gutenberg corpus (if not already downloaded)
try:
    nltk.data.find('corpora/gutenberg')
except LookupError:
    nltk.download('gutenberg')

# Load the gutenberg corpus
# We'll concatenate sentences from a few books to create a larger corpus
gutenberg_corpus_sents = []
for fileid in nltk.corpus.gutenberg.fileids()[:5]: # Use first 5 books for demonstration
    gutenberg_corpus_sents.extend(nltk.corpus.gutenberg.sents(fileid))

print(f"Loaded {len(gutenberg_corpus_sents)} sentences from the gutenberg corpus (first 5 books).")
print("First 5 sentences:")
for i, sent in enumerate(gutenberg_corpus_sents[:5]):
    print(f"  {i+1}: {sent[:10]}...") # Print first 10 words for brevity

# Prepare sentences for Word2Vec (lowercase and filter non-alphabetic tokens)
processed_sentences = [[word.lower() for word in sentence if word.isalpha()]
                       for sentence in gutenberg_corpus_sents if len(sentence) > 1]
```

### Step 2: Train Word2Vec Model on Gutenberg Corpus

Now, we'll train a new Word2Vec model using the `gutenberg` corpus. We'll use larger `vector_size` and `window` parameters to capture richer semantic relationships, and a `min_count` to ignore very rare words, making the model more robust.

```python
# Train the Word2Vec model
# vector_size: Dimensionality of the word vectors (e.g., 100-300 for larger corpora)
# window: Maximum distance between the current and predicted word within a sentence
# min_count: Ignores all words with total frequency lower than this (helps with noise)
# workers: Number of CPU cores to use for training
# sg: Training algorithm: 0 for CBOW, 1 for Skip-gram (Skip-gram is generally better for rare words)

print("Training Word2Vec model on Gutenberg corpus... This might take a few minutes.")
model_gutenberg = Word2Vec(processed_sentences,
                     vector_size=100, # Increased vector size
                     window=5,        # Increased window size
                     min_count=5,     # Ignore words appearing less than 5 times
                     workers=4,
                     sg=1)            # Using Skip-gram

print("Word2Vec model trained successfully on Gutenberg corpus!")
print(f"Vocabulary size: {len(model_gutenberg.wv.index_to_key)}")
```

## Practical Task 1: Text Prediction System

We will develop a simple text prediction system using a recurrent neural network (RNN), specifically an LSTM, to predict the next word in a sequence.

### Step 1: Import Libraries and Define Dataset

First, we'll import the necessary libraries from TensorFlow/Keras and define a small corpus for our text prediction system.

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
import numpy as np

# Define a small dataset of sentences
text_data = [
    "Natural language processing is a fascinating field",
    "NLP enables machines to understand human language",
    "Text prediction uses sequence models",
    "Deep learning revolutionizes NLP tasks",
    "Machines can learn from text data"
]

print("Text data loaded successfully.")
```

### Step 2: Preprocess the Text Data

We need to tokenize the text, create a vocabulary, and convert sentences into sequences of numbers that the neural network can process. We'll also create input sequences and their corresponding target words.

```python
# Initialize the Tokenizer
tokenizer = Tokenizer(oov_token="<unk>") # OOV for out-of-vocabulary words
tokenizer.fit_on_texts(text_data)

vocabulary_size = len(tokenizer.word_index) + 1

# Create input sequences
input_sequences = []
for line in text_data:
    token_list = tokenizer.texts_to_sequences([line])[0]
    for i in range(1, len(token_list)):
        n_gram_sequence = token_list[:i+1]
        input_sequences.append(n_gram_sequence)

print(f"Vocabulary size: {vocabulary_size}")
print(f"Total input sequences: {len(input_sequences)}")
print("First 5 input sequences:")
for seq in input_sequences[:5]:
    print(seq)
```

### Step 3: Pad Sequences and Prepare Labels

Neural networks require input sequences to have a uniform length. We will pad the sequences to the maximum length and then separate the input features (predictors) from the labels (next word).

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Find the maximum sequence length
max_sequence_len = max([len(x) for x in input_sequences])

# Pad all sequences to the same length
padded_sequences = np.array(pad_sequences(input_sequences, maxlen=max_sequence_len, padding='pre'))

# Create predictors and label
X = padded_sequences[:, :-1]
y = padded_sequences[:, -1]

# Convert labels to one-hot encoding
y = to_categorical(y, num_classes=vocabulary_size)

print(f"Max sequence length: {max_sequence_len}")
print(f"Shape of predictors (X): {X.shape}")
print(f"Shape of labels (y): {y.shape}")
print("First 5 padded input sequences (X):")
for seq in X[:5]:
    print(seq)
print("First 5 one-hot encoded labels (y):")
for label in y[:5]:
    print(label)
```

### Step 4: Build the Neural Network Model

We will construct a simple sequential model consisting of an Embedding layer (to learn word embeddings), an LSTM layer (for sequence processing), and a Dense output layer with a softmax activation for multi-class classification (predicting the next word).

```python
model = Sequential()
model.add(Embedding(vocabulary_size, 100, input_length=max_sequence_len-1)) # Embedding layer
model.add(LSTM(150)) # LSTM layer with 150 units
model.add(Dense(vocabulary_size, activation='softmax')) # Output layer with softmax activation

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.summary()
print("Model built successfully.")
```

### Step 5: Train the Model

Now, we will train our LSTM model on the prepared sequences. The model will learn to predict the next word based on the preceding words in each sequence.

```python
# Train the model
history = model.fit(X, y, epochs=100, verbose=1)

print("Model training complete.")

# Optional: plot training history
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.title('Model Accuracy Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```

### Step 6: Implement Next Word Prediction Function

We'll create a function that takes a seed text, preprocesses it, and uses the trained model to predict the most probable next word. This function can be called multiple times to generate longer text sequences.

```python
def predict_next_word(seed_text, n_words):
    for _ in range(n_words):
        token_list = tokenizer.texts_to_sequences([seed_text])[0]
        token_list = pad_sequences([token_list], maxlen=max_sequence_len-1, padding='pre')
        predicted = np.argmax(model.predict(token_list, verbose=0), axis=-1)

        output_word = ""
        for word, index in tokenizer.word_index.items():
            if index == predicted:
                output_word = word
                break
        seed_text += " " + output_word
    return seed_text

print("Next word prediction function defined.")
```

```python
import numpy as np
from tensorflow.keras.preprocessing.sequence import pad_sequences

def predict_next_word_with_confidence(seed_text, num_predictions=3):
    """
    Predicts the top N next word(s) and their confidence for a given seed text.
    """
    original_seed_text = seed_text

    # Tokenize the seed text. texts_to_sequences returns a list of sequences, we take the first one.
    token_list = tokenizer.texts_to_sequences([seed_text])[0]

    if not token_list:
        print(f"\nSeed Text: '{original_seed_text}'")
        print("  Could not tokenize the input text. Please use words from the training vocabulary.")
        return

    # Pad the token list to the length the model expects (max_sequence_len - 1)
    # padding='pre' matches how the training data was prepared
    padded_token_list = pad_sequences([token_list], maxlen=max_sequence_len - 1, padding='pre')

    # Get prediction probabilities from the model
    # model.predict returns a numpy array, we take the first (and only) prediction vector
    predictions = model.predict(padded_token_list, verbose=0)[0]

    # Get the indices of the top N most probable words
    # argsort() gives indices that would sort the array. Slicing with [-num_predictions:] gets the top N
    # [::-1] reverses it to get highest confidence first
    top_n_indices = predictions.argsort()[-num_predictions:][::-1]

    print(f"\nSeed Text: '{original_seed_text}'")
    print(f"Predicted Next Word(s) with Confidence:")

    found_predictions = 0
    for idx in top_n_indices:
        # Check if the index corresponds to a known word in the tokenizer's vocabulary
        # tokenizer.index_word maps integer indices back to words
        word = tokenizer.index_word.get(idx)

        # If the word is the OOV token itself and it's not explicitly in index_word, skip or handle.
        # Our tokenizer was initialized with oov_token="<unk>", which should have an entry in index_word.
        if word is None:
            continue

        confidence = predictions[idx]
        print(f"  - '{word}' (Confidence: {confidence:.4f})")
        found_predictions += 1

    if found_predictions == 0:
        print("  No valid predictions found for this seed text.")


# Interactive loop for the mini predictive text application
print("=" * 60)
print("  MINI PREDICTIVE TEXT APPLICATION")
print("=" * 60)
print("Enter a sentence fragment to get next word predictions.")
print("Type 'exit' to quit.")

while True:
    user_input = input("\nEnter your fragment: ").strip()
    if user_input.lower() == 'exit':
        print("Exiting application. Goodbye!")
        break
    if not user_input:
        print("Please enter some text.")
        continue

    # Predict the top 3 most probable words by default
    predict_next_word_with_confidence(user_input, num_predictions=3)
```

## Practical Task 2: Text Generation System with GPT-2

In this task, we will build a simple text generation application using the GPT-2 model from the Hugging Face Transformers library. This system will accept user input as a prompt and generate creative text continuations.

### Step 1: Install and Import Libraries

We need the `transformers` library to access pre-trained GPT-2 models and tokenizers. (Note: `transformers` was already installed in a previous task, so this cell ensures it's available and imports the necessary modules.)

```python
# The transformers library might already be installed, but we ensure it is.
!pip install transformers --quiet

from transformers import pipeline, set_seed

print("Libraries imported successfully.")
```

### Step 2: Load the GPT-2 Model and Tokenizer

Hugging Face provides a convenient `pipeline` function for various NLP tasks, including text generation. We'll load the `gpt2` model for this purpose.

```python
# Set a seed for reproducibility
set_seed(42)

# Initialize the text-generation pipeline with GPT-2
generator = pipeline('text-generation', model='gpt2')

print("GPT-2 model and tokenizer loaded successfully.")
```

### Step 3: Implement Text Generation Function

Now, let's create a function that takes a prompt and uses the loaded GPT-2 model to generate text continuations. We'll allow parameters for `max_length` to control the length of the generated text and `num_return_sequences` to get multiple diverse outputs.

```python
def generate_text_with_gpt2(prompt, max_length=50, num_return_sequences=1):
    """
    Generates text continuations for a given prompt using the GPT-2 model.

    Args:
        prompt (str): The initial text to continue.
        max_length (int): The maximum number of tokens in the generated text (including the prompt).
        num_return_sequences (int): The number of independent sequences to generate.

    Returns:
        list: A list of generated text strings.
    """
    generated_outputs = generator(
        prompt,
        max_length=max_length,
        num_return_sequences=num_return_sequences,
        truncation=True # Truncate input if it's too long for the model
    )
    return [output['generated_text'] for output in generated_outputs]

print("Text generation function defined.")
```

### Step 4: Interactive Text Generation

Finally, let's create an interactive loop where you can enter different prompts, and the GPT-2 model will generate continuations. You can experiment with various prompts and observe how the model responds.

```python
print("=" * 70)
print("       GPT-2 INTERACTIVE TEXT GENERATION")
print("=" * 70)
print("Enter a prompt sentence to generate text continuations.")
print("Type 'exit' or 'quit' to end the application.")
print("You can also specify max_length (e.g., 'prompt, max_length=70') or num_sequences (e.g., 'prompt, num_sequences=3').")

while True:
    try:
        user_input = input("\nEnter your prompt: ").strip()

        if user_input.lower() in ['exit', 'quit']:
            print("Exiting text generation application. Goodbye!")
            break

        if not user_input:
            print("Please enter some text.")
            continue

        # Parse user input for optional parameters
        prompt_parts = user_input.split(', ')
        prompt = prompt_parts[0]
        max_len = 50
        num_seq = 1

        for part in prompt_parts[1:]:
            if part.startswith('max_length=') or part.startswith('max_len='):
                try:
                    max_len = int(part.split('=')[1])
                except ValueError:
                    print("Invalid max_length. Using default (50).")
            elif part.startswith('num_sequences=') or part.startswith('num_seq='):
                try:
                    num_seq = int(part.split('=')[1])
                except ValueError:
                    print("Invalid num_sequences. Using default (1).")

        generated_texts = generate_text_with_gpt2(prompt, max_length=max_len, num_return_sequences=num_seq)

        print("\n--- Generated Texts ---")
        for i, text in enumerate(generated_texts):
            print(f"Sequence {i+1}: {text}\n")

    except Exception as e:
        print(f"An error occurred: {e}")
        print("Please try again or type 'exit' to quit.")
```
